# stats -- A pickleball stats library 

> A simple API for using and analyzing pickleball statistics

```python
#| hide
from nbdev.showdoc import *
from fastcore.test import *
import pandas as pd
```

```python
#| export
table_names = ["tournament", "match"]
```

These are the names of the tables that exist in the pklmart database

```python
#| export
class Table:
    "A representation of a table in the pklmart database"
    def __init__(self, 
                table_name:str # The name of the table
                ):
        if not isinstance(table_name, str):
            raise TypeError(f"table_name must be a string within {table_names}")
        if table_name not in table_names:
            raise ValueError(f"Table name {table_name} is not a name in table_names")
        self.table_name = table_name
        self._df = None
    def __str__(self): return f"{self.table_name} table"
    __repr__ = __str__
```

```python
show_doc(Table)
```

Here's an example of creating and displaying a Table object

```python
t = Table("match")
t
```

Ensure the table name is the correct type and is in the database

```python
test_fail(lambda: Table(1), contains="table_name must be a string within")
test_fail(lambda: Table("math"), contains="Table name math is not a name in table_names")
```
